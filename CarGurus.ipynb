```python
#Brand value
brands = []
driver = webdriver.Chrome(ChromeDriverManager().install())
url = 'https://www.cargurus.com/Cars/inventorylisting/viewDetailsFilterViewInventoryListing.action?zip=23102&showNegotiable=true&sourceContext=forSaleTab_false_0&distance=500&entitySelectingHelper.selectedEntity=m3'
driver.get(url)
pageSource = driver.page_source
bs = BeautifulSoup(pageSource, 'html.parser')
for element in bs.find("optgroup", {'label':'All Makes'}).find_all('option'):
    brands.append(element.attrs['value'])
#Brand dataframe
car = []
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://www.cargurus.com/Cars/inventorylisting/viewDetailsFilterViewInventoryListing.action?zip=23102&showNegotiable=true&sourceContext=forSaleTab_false_0&distance=500&entitySelectingHelper.selectedEntity=m3#resultsPage=2")
pageSource = driver.page_source
bs = BeautifulSoup(pageSource, 'html.parser')
for element in bs.find("optgroup", {'label':'All Makes'}).find_all('option'):
    car.append(element.text)
car_df = pd.DataFrame(car, columns=['Name'])
zip_code = ['97884', '58568', '60047', '06371', '89067', '79019', '75935', '31313']
car_df['97884'] = 0
car_df['58568'] = 0
car_df['60047'] = 0
car_df['06371'] = 0
car_df['89067'] = 0
car_df['79019'] = 0
car_df['75935'] = 0
car_df['31313'] = 0
car_df['value'] = brands
car_df = car_df.set_index('value')
#Web pages
for value in brands:
    for number in zip_code:
        url = "https://www.cargurus.com/Cars/inventorylisting/viewDetailsFilterViewInventoryListing.action?zip="+str(number)
        +"&showNegotiable=true&sourceContext=forSaleTab_false_0&distance=500&entitySelectingHelper.selectedEntity="+str(value)
        driver.get(url)
        pageSource = driver.page_source
        bs = BeautifulSoup(pageSource, 'html.parser')
        test = bs.find("div",class_='VmQTCB')
        pages = test.get_text()
        pages = math.ceil(int(pages.split(' ')[0])/15)
        car_df = car_df.set_value(value, number, pages)
```

```python
#Item List
start = time.time()
zip_code = ['97884', '58568', '60047', '06371', '89067', '79019', '75935', '31313']
columns = ["Title:" ,"Location:", 'Gas Mileage:', "Dealer's Price:", "Mileage:",  'Transmission:', "Exterior Color:",
           "Interior Color:", "Maximum Seating:", "Engine:", "Drivetrain:", "VIN:",'Stock #:','Major Options:', 
           'Certified:', 'Title Check', 'Accident Check', 'Ownership history', "Days on CarGurus", "Price History"]
Listing_Summary = pd.DataFrame(columns=columns)
driver = webdriver.Chrome(ChromeDriverManager().install())
for brand_value in ['m3']:
    for code_value in ['97884']:
        for i in range(100):
            url = "https://www.cargurus.com/Cars/inventorylisting/viewDetailsFilterViewInventoryListing.action?zip="+str(code_value)+"&showNegotiable=true&sourceContext=forSaleTab_false_0&distance=500&entitySelectingHelper.selectedEntity="+str(brand_value)+"#resultsPage="+str(i)
            time.sleep(0.1)
            driver.get(url)
            pageSource = driver.page_source
            bs = BeautifulSoup(pageSource, 'html.parser')
            item_list = bs.find("div", attrs={'data-cg-ft':'srp-all-listing-blades'}).select("div[class='SimGDW']")
            for item in item_list:
                url_item = "https://www.cargurus.com/Cars/inventorylisting/viewDetailsFilterViewInventoryListing.action?zip="+str(code_value)+"&showNegotiable=true&sourceContext=forSaleTab_false_0&distance=500&entitySelectingHelper.selectedEntity="+str(brand_value)+str(item.find('a').get('href'))
                driver.get(url_item)
                time.sleep(0.1)
                Pagesource = driver.page_source
                bs_item = BeautifulSoup(Pagesource, 'html.parser')
                summary_item = bs_item.select("dd[class='_5grpKY undefined']")
                if summary_item == []:
                    continue
                item_title = bs_item.select("div[class='_2Nz9KW']")
                summary_name = bs_item.select("dt[class='aHpS63 undefined']")
                name_list = []
                for i in summary_name:
                    name_list.append(i.string)
                    
                test_df = pd.DataFrame(columns=name_list)
                item_detail = []
            #special case
                if 'Location:' in name_list:
                    index_loc = name_list.index('Location:')
                    for i in summary_item[index_loc].select('span')[0]:
                        item_detail.insert(0, i.string)
                        break
                    summary_item.pop(index_loc)
                    name_list.pop(index_loc)
                    name_list.insert(0, 'Location:')
            #special case
                if 'Gas Mileage:' in name_list:
                    index_mile = name_list.index('Gas Mileage:')
                    for i in summary_item[index_mile-1].select('p')[0:2]:
                        item_detail.insert(1, i.string)
                    item_detail[1] = item_detail[1]+', '+item_detail.pop(2)
                    summary_item.pop(index_mile-1)
                    name_list.pop(index_mile)
                    name_list.insert(1,'Gas Mileage:')
            #main append          
                for item in summary_item: 
                    item_detail.append(item.string)
                name_list.insert(0, 'Title:')
                item_detail.insert(0, item_title[0].string)
            #vehicle history
                accident_bs = bs_item.select("span[class='_5GJRd3']")
                for i in accident_bs[0:2]:
                    name_list.append(i.string)
                name_list.append('Ownership history')
                title_detail = bs_item.select("span[class='_5gudF3']")
                item_detail.append(title_detail[0].string)
                accident_detail = bs_item.select("div[class=_5gudF3]")
                item_detail.append(accident_detail[0].select('span')[0].string)
                item_detail.append(accident_bs[-1].string)
            #Negotitaion
                name_list.append('Days on CarGurus')
                name_list.append('Price History')
            
            #
                price_history = bs_item.select("div[class='_3ryQpg']")
                if len(price_history) == 2:
                    item_detail.append(price_history[0].select('strong')[0].string)
                    item_detail.append(price_history[1].string)
                elif len(price_history) == 1:
                    days_list = bs_item.select("div[class='_5XcXHD']")
                    item_detail.append(days_list[0].select('strong')[0].string)
                    item_detail.append(price_history[0].string)
                elif len(price_history) == 3:
                    item_detail.append(price_history[1].select('strong')[0].string)
                    item_detail.append(price_history[2].string)
                print(name_list)
                print(item_detail)
            #concatenate data
                data_set = []
                data_set.append(item_detail)
                df_gurus = pd.DataFrame(data=data_set, columns=name_list)
                df_gurus
                Listing_Summary = pd.concat([Listing_Summary, df_gurus], sort=False, ignore_index=True)
Listing_Summary
Total_time = (time.time()-start)
Total_time
```
